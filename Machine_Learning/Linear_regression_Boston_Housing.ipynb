## Step 1: Load data & assigning them to X and Y matrices

```python
## Loading dataset:

from sklearn.datasets import load_boston
import pandas as pd
```

```python
## Loading dataset
data = load_boston()
```

```python
data.keys()
```

### Data description

```python
print(data.DESCR)
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
df['target'] = pd.Series(data.target)
df
```

```python
df.describe()
```

## Creating X & Y Data matrices :

```python
## Creating X & Y matrices :
X = data.data
Y = data.target
```

```python
X.shape ,  Y.shape
# X contains 442 rows and 10 columns , Y contains 442 rows & 1 column
```

```python
## Step 1 can be done inn one simple method:
X,Y = load_boston(return_X_y=True)
X.shape , Y.shape
```

## Setp 2 : Splitting Data to train & split (80/20 Split)

```python
## 1. importing libraries :
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X, Y ,test_size = 0.2) ## 80% train, 20% test
```

```python
# using random seed to shuffle the data by using random_state
X_train , X_test , Y_train , Y_test = train_test_split(X, Y, random_state = 4,test_size = 0.2)
```

```python
## Data shape : ** num rows has to be same allways 
X_train.shape,Y_train.shape
```

```python
X_test.shape , Y_test.shape
```

## Step 3: Build Linear Regression Model

```python
## Importing libraries :
from sklearn import linear_model ## model library 
from sklearn.metrics import mean_squared_error,r2_score ## testing performance of our model, metrics which are required
```

```python
model = linear_model.LinearRegression()
model
```

```python
## training linear regression model on tranning data
model.fit(X_train,Y_train)
```

```python
## using model to predict on the testing data i.e X_test
Y_pred = model.predict(X_test)
```

## Comapring prediction with actual value

```python
## Predicted value
Y_pred[1:5]
```

```python
## Actual value
Y_test[1:5]
```

## Step 4 : Evaluating model performance

```python
## Step 4 : Evaluating model performance
print("Coefficients:",model.coef_) ## This represnt the weight values of the 10 features(variable)
print("Intercept:",model.intercept_)
## %.2f is used for string formatting using modulo 
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(Y_test, Y_pred))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Y_test, Y_pred))
```
