```python
import torch
import torch.nn.functional as functional
```

```python
## creating data:

x = torch.randn(100,1) ##this will create a torch of size 200 ,1 using func random num generator 
# we are creating y such that it contains a function plus some noise of size of input
y = x.pow(3) + 0.5*torch.randn(x.size())
```

# 1. Implementing manually:


```python
epoch = 150
learning_rate = 0.05
```

```python
## we are making our own neural network class over torch neural network module:

''' 
num_inputs =  Number of inputs is just total number of features

'''
class Net(torch.nn.Module):
    def __init__(self,n_inputs,n_hidden,n_output):
        ## Calling constructor of super class:
        super(Net,self).__init__() ## calling contructor of nn module
        
        '''nn.Linear - take in two arguments -  
        size of input layer & size of output layer'''
        
        self.hidden = torch.nn.Linear(n_inputs,n_hidden) ## creates a fully connected linear layer - 
        
        self.output = torch.nn.Linear(n_hidden,n_output)
        

    ## CREATING FORWAARD PROPAGATING FUNCTION: 
    
    def forward(self,input):
        ## We already have the input layer , we will calc hidden layer for forward propagation
        hidden = functional.relu(self.hidden(input)) ## Activation unit
        output = self.output(hidden)
        
        return output
        
        
```

```python
## creating objett:

net = Net(n_inputs=1,n_hidden=20,n_output=1)
```

```python
print(net)
```

```python
## Creating loss function: here we are using mean sure error:

loss_func = torch.nn.MSELoss()
```

```python
## now we will run this : this is optimization function using gradient descent 

for i in range(epoch):
    output = net(x)
    loss = loss_func(output,y)
    ## zero grad will clear the buffer which was calculated for forward pass using gradient descent to zero
    net.zero_grad()
    loss.backward() ## backward propagation
    ## updating values - there are various optimizer liek AdamW etc 
    
    ## parameters in model function which is net , gives acces to all the model parameters i.e weights biases 
    for f in net.parameters():
        
        ## to acess the value:
        f.data.sub_(learning_rate*f.grad.data) ## __ is used for pass by reference & sub will subsract the value in f 
    
    print(i, 'th loss ',loss.squeeze())
```

## 2. Building Simple neural network architecture : using - sequential

```python
seq_net = torch.nn.Sequential(torch.nn.Linear(1,20), ## input layer - linear layer , 
                             torch.nn.ReLU(), ## activation function
                             torch.nn.Linear(20,1)) ## ouput layer which is again the linear layer 
```

```python
## This is same is manually built network architecture 
print(f'Manually build layer : {net}')
print(f'using sequential layer function : {seq_net}')
```

```python
loss_func = torch.nn.MSELoss()
optim = torch.optim.SGD(seq_net.parameters(),lr=0.05,momentum=0.09)
```

```python
import matplotlib.pyplot as plt
import numpy as np

## now we will run this : this is optimization function using gradient descent 

for i in range(epoch):
    output = seq_net(x)
    loss = loss_func(output,y)
    ## zero grad will clear the buffer which was calculated for forward pass using gradient descent to zero
    optim.zero_grad()
    loss.backward() ## backward propagation
    ## updating values - there are various optimizer liek AdamW etc 
    
    optim.step() # this will update values in each setp
    
    # printing loss in every step:
    if (i+1) %2 ==0: ## printing even loss
        print(str(i+1)+'th loss', loss.squeeze())
        
    ## Plotting learning 
    
    if i %5==0:
        plt.cla()
        plt.scatter(x.data.numpy(),y.data.numpy())
        plt.plot(x.data.numpy(),output.data.numpy(),'r-',lw=5)
#         plt.text(0.5,0,'Loss=%.4f' % loss.data[0],fontdict={'size':20,'color': 'red'})
        plt.pause(0.1)
plt.ioff()
plt.show()
```
